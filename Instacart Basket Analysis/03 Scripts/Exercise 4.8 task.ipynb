# Part 4

```python
# Using transform() to create max_order column
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Creating loyatly flag using loc()
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking to see that loyalty_flag column is displayed
ords_prods_merged.head()
```

# Part 5

```python
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Loyal customers have the lowest aggregate mean price where new customers have the highest

# Part 6

```python
# making new column for mean_price
ords_prods_merged['mean_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# creating spending flag
```

```python
ords_prods_merged.loc[ords_prods_merged['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# checking to see if new columns are there
ords_prods_merged.head()
```

# Part 7 

```python
# making new column for median days since prior order
ords_prods_merged['median_days_since_prior_order'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Creating flag for order frequency 
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_days_since_prior_order'] > 10) & (ords_prods_merged['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Checking outputs of new columns 
ords_prods_merged.head()
```
