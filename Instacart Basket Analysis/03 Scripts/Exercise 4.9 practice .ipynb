```python
# Importing Libraries 
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_update.pkl'))
```

## Creating Bar Charts 

```python
# create a bar chart 
ords_prods_merged['orders_day_of_week'].value_counts().plot.bar()
```

```python
# sort by index number and add color 
ords_prods_merged['orders_day_of_week'].value_counts().sort_index().plot.bar(color = ['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Exporting bar chart 
bar = ords_prods_merged['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
# Saving figure outside of Jupyter 
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

## Creating Histograms and Scatterplots

```python
# Making a histogram
ords_prods_merged['prices'].plot.hist(bins = 25)
```

```python
# Making a scatterplot 
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_merged)
```

```python
# Checking for outliers
ords_prods_merged.loc[ords_prods_merged['prices'] > 100]
```

```python
ords_prods_merged.loc[ords_prods_merged['prices'] >100, 'prices'] = np.nan
```

```python
# checking for new max value 
ords_prods_merged['prices'].max()
```

```python
hist = ords_prods_merged['prices'].plot.hist(bins = 25)
```

```python
# Saving and exporting hist
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
```
