# Part 3 

```python
# Creating a histogram
hist = df_ords_prods_customers['order_hour_of_day'].plot.hist(bins = 24)
```

The histogram shows the highest frequency of order_hour_of_day happens between hour 10 and 16. The lowest frequency occurs between hours 1 and 5. 

```python
# Saving figure as png
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_order_hour_of_day.png'))
```
